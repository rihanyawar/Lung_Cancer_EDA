```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("Lung Cancer.csv")
df.head()
```

```python
df.info()
```

```python
#Summary of the dataframe
df.describe()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
#Separating the date column values because the model will not understand it

df['diagnosis_date'] = pd.to_datetime(df['diagnosis_date'])

df['diagnosis_YEAR'] = df['diagnosis_date'].dt.year
df['diagnosis_MONTH'] = df['diagnosis_date'].dt.month
df['diagnosis_DATE'] = df['diagnosis_date'].dt.day
```

```python
#Now delete the diagnosis_date column
df.drop('diagnosis_date',axis=1, inplace=True)
```

```python
df.head()
```

```python
#Now convert the end_treatment_date in proper format as that of diagnosis_date
df['end_treatment_date']=pd.to_datetime(df['end_treatment_date'])

df['end_treatment_YEAR'] = df['end_treatment_date'].dt.year
df['end_treatment_MONTH'] = df['end_treatment_date'].dt.month
df['end_treatment_DATE'] = df['end_treatment_date'].dt.day

#Now drop the end_treatment_date column
df.drop('end_treatment_date',axis=1, inplace=True)
```

```python
df.head()
```

```python
#Now distribute the data according to the gender
plt.figure(figsize=(6, 5))
sns.countplot(x="gender",data=df, palette='viridis')
plt.title("Distribution o Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Distribution of data on the basis of age
plt.figure(figsize=(10,6))
sns.histplot(df['age'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#Country wise distribution
df['country'].value_counts().plot(kind="bar",color="skyblue")
plt.title("Country Wise Distribution")
plt.xlabel("Country")
plt.ylabel("Counts")
plt.show()
```

```python
#Countires involved
df['country'].unique()
```

```python
# Get counts of each smoking status category
smoking_counts = df["smoking_status"].value_counts()

# Plot
plt.figure(figsize=(12, 5))

# First subplot - Bar chart
plt.subplot(1, 2, 1)
sns.countplot(x='smoking_status', data=df, palette='plasma')
plt.title('Count of Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.xticks(rotation=30)

# Second subplot - Pie chart
plt.subplot(1, 2, 2)
plt.pie(smoking_counts, labels=smoking_counts.index, autopct="%1.1f%%", shadow=True)
plt.title('Percentage of Smoking Status')

plt.tight_layout()
plt.show()
```

```python
df['smoking_status'].value_counts()
```

```python
#Distribution of people survived
plt.figure(figsize=(6, 5))
sns.countplot(x='survived', data=df, palette='rocket')
plt.title('Distribution of Survival Status (0=No, 1=Yes)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
#Bivariate Analysis
#Relationship between SMOKING_STATUS and SURVIVED

plt.figure(figsize=(10, 6))
sns.countplot(x='smoking_status', hue='survived', data=df, palette='viridis')
plt.title('Survival Status by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

pd.crosstab(df['smoking_status'], df['survived'])
```

```python
#Bivariate Analysis
#Relationship between cancer_stage and survived

plt.figure(figsize=(10, 6))
sns.countplot(x='cancer_stage', hue='survived', data=df, palette='plasma', order=df['cancer_stage'].value_counts().index)
plt.title('Survival Status by Cancer Stage')
plt.xlabel('Cancer Stage')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

pd.crosstab(df['cancer_stage'],df['survived'])
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='family_history', hue='survived', data=df, palette='rocket')
plt.title('Survival Status by Family History (0=No, 1=Yes)')
plt.xlabel('Family History')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

pd.crosstab(df['family_history'],df['survived'])
```

```python
#Bivariate Analysis
#Relationship between age and survived

sns.boxplot(x='survived',y='age',data=df,palette='Set2')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survived (0=No, 1=Yes)')
plt.ylabel('Age')
plt.show()
```

```python
#Bivariate Analysis
#Relationship between bmi and survived

plt.figure(figsize=(7, 6))
sns.boxplot(x='survived', y='bmi', data=df, palette='Set2')
plt.title('BMI Distribution by Survival Status')
plt.xlabel('Survived (0=No, 1=Yes)')
plt.ylabel('BMI')
plt.show()
```

```python
#Visualize the correlation matrix with a Heatmap

plt.figure(figsize=(12, 10))
sns.heatmap(df[['cholesterol_level','age','bmi','hypertension','asthma','cholesterol_level','cirrhosis','other_cancer']].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
#Number of diagnoses per year
sns.countplot(x='diagnosis_YEAR', data=df, palette='crest')
plt.title('Number of Lung Cancer Diagnoses Per Year')
plt.xlabel('Diagnosis Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
